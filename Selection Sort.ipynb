```python
def selection_sort_ascend(x):
    for item in x:
        temp = item
        start_index = x.index(temp)
        final_index = start_index
        for num,j in enumerate(x[start_index:]):
            if(j <= temp):
                temp = j
                final_index = num + start_index
        temp_index = x[start_index]
        x[start_index] = x[final_index]
        x[final_index] = temp_index
    return x
```

```python
t1 = time.time()
```

```python
def selection_sort_descend(x):
    for item in x:
        temp = item
        start_index = x.index(temp)
        final_index = start_index
        for num,j in enumerate(x[start_index:]):
            if(j >= temp):
                temp = j
                final_index = num + start_index
        temp_index = x[start_index]
        x[start_index] = x[final_index]
        x[final_index] = temp_index
    return x
```

```python
t1 = time.time()
```

```python
selection_sort_ascend(data)
```

```python
selection_sort_descend(data)
```
