```python
def bubble_sort_ascend(x):
    for i in range(0,len(x)-1):
        for j in range(0,len(x)-1):
            if(x[j] > x[j+1]):
                x[j],x[j+1] = x[j+1],x[j]
    return x
```

```python
def bubble_sort_decend(x):
    for i in range(0,len(x)-1):
        for j in range(0,len(x)-1):
            if(x[j] > x[j+1]):
                x[j],x[j+1] = x[j+1],x[j]
    return x
```

```python
y = [5,5,5,5,4,4,4,4,3,3,3,3,2,2,2,2]
```

```python
bubble_sort_ascend(y)
```

```python
bubble_sort_decend(y)
```
